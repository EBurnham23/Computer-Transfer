```python
import numpy as np #importing our packages to unpack the data file and plot
from matplotlib import pyplot as plt

#----half----#

#We skip the first line since they are just labels

columns_half = np.loadtxt('half_history.data', skiprows = 1, unpack = True)
L_surf_half = columns_half[33] #linear surface luminosity
T_eff_half = columns_half[34] #effective temperature in log
R_half = columns_half[37] #log radius
Age_half = columns_half[1] #age of star

#----one----#

#We skip the first line since they are just labels

columns_one = np.loadtxt('one_history.data', skiprows = 1, unpack = True)
L_surf_one = columns_one[33] #linear surface luminosity
T_eff_one = columns_one[34] #effective temperature in log
R_one = columns_one[37] #log radius
Age_one = columns_one[1] #age of star

#----five----#

#We skip the first line since they are just labels

columns_five = np.loadtxt('five_history.data', skiprows = 1, unpack = True)
L_surf_five = columns_five[33] #linear surface luminosity
T_eff_five = columns_five[34] #effective temperature in log
R_five = columns_five[37] #log radius
Age_five = columns_five[1] #age of star

#----fifteen----#

#We skip the first line since they are just labels

columns_fifteen = np.loadtxt('fifteen_history.data', skiprows = 1, unpack = True)
L_surf_fifteen = columns_fifteen[33] #linear surface luminosity
T_eff_fifteen = columns_fifteen[34] #effective temperature in log
R_fifteen = columns_fifteen[37] #log radius
Age_fifteen = columns_fifteen[1] #age of star

#----fifty----#

#We skip the first line since they are just labels

columns_fifty = np.loadtxt('fifty_history.data', skiprows = 1, unpack = True)
L_surf_fifty = columns_fifty[33] #linear surface luminosity
T_eff_fifty = columns_fifty[34] #effective temperature in log
R_fifty = columns_fifty[37] #log radius
Age_fifty = columns_fifty[1] #age of star

#----graphing----#

fig, ax =  plt.subplots(nrows=2, ncols=1, figsize=(20, 30))

ax[0].set_title('HR Diagram', fontsize = 20)
ax[0].scatter(T_eff_half, L_surf_half, c= 'red', label = r'$\frac{1}{2} M_{\odot}$')
ax[0].scatter(T_eff_one, L_surf_one,  c = 'orange', label = r'$ M_{\odot}$')
ax[0].scatter(T_eff_five, L_surf_five, c = 'yellow', label = r'$ 5M_{\odot}$')
ax[0].scatter(T_eff_fifteen, L_surf_fifteen,  c = 'green', label = r'$ 15M_{\odot}$')
ax[0].scatter(T_eff_fifty, L_surf_fifty,  c = 'blue', label = r'$50 M_{\odot}$')
ax[0].invert_xaxis() #inverting temperature axis
ax[0].set_xlabel(r"Temperature (Log) [ $\frac{T_{eff}}{K} $]", fontsize = 15)
ax[0].set_ylabel(r"Luminosity (Log) [$\frac{L_{surface}}{L_{\odot}}$]", fontsize = 15)
ax[0].legend()

#creating labels to indicate final positions

ax[0].text(3.8,-18, 'Initial', style='italic', fontsize=12,
        bbox={'facecolor': 'grey', 'alpha': 0.5, 'pad': 10})
ax[0].text(4.3,6, 'Final', style='italic', fontsize=12,
        bbox={'facecolor': 'lightblue', 'alpha': 0.5, 'pad': 10})
ax[0].text(3.59,0, 'Final', style='italic', fontsize=12, 
        bbox={'facecolor': 'salmon', 'alpha': 0.5, 'pad': 10})
ax[0].text(3.75,1.1, 'Final', style='italic', fontsize=12, 
        bbox={'facecolor': 'sandybrown', 'alpha': 0.5, 'pad': 10})
ax[0].text(4.08,3, 'Final', style='italic', fontsize=12, 
        bbox={'facecolor': 'lemonchiffon', 'alpha': 0.5, 'pad': 10})
ax[0].text(4.35,4, 'Final', style='italic', fontsize=12, 
        bbox={'facecolor': 'lightgreen', 'alpha': 0.5, 'pad': 10})

ax[1].set_title('Luminosity v. Radius', fontsize = 20)
ax[1].scatter(R_half, L_surf_half, c = 'red', label = r'$\frac{1}{2} M_{\odot}$', s = 40)
ax[1].scatter(R_one, L_surf_one, c = 'orange', label = r'$ M_{\odot}$', s = 40)
ax[1].scatter(R_five, L_surf_five, c = 'yellow', label = r'$ 5M_{\odot}$', s = 40)
ax[1].scatter(R_fifteen, L_surf_fifteen, c = 'green', label = r'$ 15M_{\odot}$', s = 40)
ax[1].scatter(R_fifty, L_surf_fifty, c = 'blue', label = r'$50 M_{\odot}$', s = 40)
ax[1].set_xlabel(r"Surface Radius [ $log(\frac{R}{R_{\odot}}) $]", fontsize = 15)
ax[1].set_ylabel(r"Surface Luminosity (Linear) [$\frac{L_{surface}}{L_{\odot}}$]", fontsize = 15)
ax[1].legend()

ax[1].text(0.9,-20, 'Initial', style='italic', fontsize=12,
        bbox={'facecolor': 'salmon', 'alpha': 0.5, 'pad': 10})
ax[1].text(1.53,-20, 'Initial', style='italic', fontsize=12,
        bbox={'facecolor': 'sandybrown', 'alpha': 0.5, 'pad': 10})
ax[1].text(1.8,-21, 'Initial', style='italic', fontsize=12,
        bbox={'facecolor': 'lemonchiffon', 'alpha': 0.5, 'pad': 10})
ax[1].text(2.7,-21, 'Initial', style='italic', fontsize=12,
        bbox={'facecolor': 'lightgreen', 'alpha': 0.5, 'pad': 10})
ax[1].text(2.7,-15, 'Initial', style='italic', fontsize=12,
        bbox={'facecolor': 'lightblue', 'alpha': 0.5, 'pad': 10})


ax[1].text(1.9,5.5, 'Final', style='italic', fontsize=12,
        bbox={'facecolor': 'lightblue', 'alpha': 0.5, 'pad': 10})
ax[1].text(-.3,0, 'Final', style='italic', fontsize=12, 
        bbox={'facecolor': 'salmon', 'alpha': 0.5, 'pad': 10})
ax[1].text(.1,1, 'Final', style='italic', fontsize=12, 
        bbox={'facecolor': 'sandybrown', 'alpha': 0.5, 'pad': 10})
ax[1].text(.4,4, 'Final', style='italic', fontsize=12, 
        bbox={'facecolor': 'lemonchiffon', 'alpha': 0.5, 'pad': 10})
ax[1].text(0.8,5.5, 'Final', style='italic', fontsize=12, 
        bbox={'facecolor': 'lightgreen', 'alpha': 0.5, 'pad': 10})


plt.show()
```
